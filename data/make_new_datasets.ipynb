```python
import pandas as pd
```

```python
full_data = pd.read_csv("additional_data/53k_parse_Labse_data.csv").query('score_orig >= 30 & score_orig < 95')
old_train = pd.read_csv("cleared_v2/cleared_v2_train_09.csv").rename(columns={'ru': 'rus'}).drop(columns=['ind'])
old_val = pd.read_csv('cleared_v2/cleared_v2_val_005.csv').rename(columns={'ru': 'rus'}).drop(columns=['ind'])
old_test = pd.read_csv('cleared_v2/cleared_v2_test_005.csv').rename(columns={'ru': 'rus'}).drop(columns=['ind'])
```

```python
replace_table_default = {
    '\uf519': 'о̄',
    '\uf523': 'э̄',
    '\uf50f': 'а̄',
    '\uf511': 'ē',
    '\uf529': 'а̄',
    '\uf518': 'о̄',
    '\uf528': 'Я̄',
    '\uf513': 'ё̄',
    '\uf522': 'Э̄',
    '\uf512': 'Ё̄',
    '\uf50e': 'А̄',
    '\uf521': 'ы̄',
    '\uf52d': 'ю̄',
    '\uf52c': 'Ю̄',
    '\uf510': 'Ē',
    'ū': 'ӣ',
    '\t': ' ',
    '\r': ' ',
    '\xa0': ' ',
    '\xad': '',
    '\\': ''
}

replace_table_macrons = {
    'ā': 'а̄',
    'Ā': 'А̄',
    'ӯ': 'ӯ',
    'Ӯ': 'Ӯ',
    'ȳ': 'ӯ',
    'Ȳ': 'Ӯ',
    'ō': 'о̄',
    'Ō': 'О̄',
    'ē': 'е̄',
    'Ē': 'Е̄',
    'ӣ': 'ӣ',
    'Ӣ': 'Ӣ',
    'ă': 'а̄',
    'á': 'а̄'
}
```

```python
def full_data_preproc(df_orig: pd.DataFrame, change_macrons=False, cols2change=['rus', 'mansi']) -> pd.DataFrame:
    df = df_orig.copy()
    for col in cols2change:
        # замена символов, как при формировании трейна (еще давно)
        for old, new in replace_table_default.items():
            df[col] = df[col].apply(lambda x: x.replace(old, new))
            
        # замена символов на кириллицу + макрон
        if change_macrons:
            for old, new in replace_table_macrons.items():
                df[col] = df[col].apply(lambda x: x.replace(old, new))
                
        # удаление лишних пробелов
        df[col] = df[col].apply(lambda x: " ".join(x.split()))
    return df
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
small_data = full_data[full_data.score_orig > 60].copy().reset_index(drop=True)
```

# 1. Все новые данные, без изменения макронов

```python
full_train_wout_macrons = pd.concat([
    full_data[['rus', 'mansi']],
    old_train[['rus', 'mansi']]
], axis=0, ignore_index=True).copy()

full_train_wout_macrons = full_data_preproc(
    full_train_wout_macrons,
    change_macrons=False,
    cols2change=['rus', 'mansi']
)

full_train_wout_macrons['split'] = 'train'
full_train_wout_macrons = full_train_wout_macrons[old_val.columns]
full_train_wout_macrons
```

```python
full_train_wout_macrons.to_csv('cleared_v3.1/big_orig_preproc/cleared_v3_train_09.csv', index=False)
old_val.to_csv('cleared_v3.1/big_orig_preproc/cleared_v3_val_005.csv', index=False)
old_test.to_csv('cleared_v3.1/big_orig_preproc/cleared_v3_test_005.csv', index=False)
```

# 2. Все новые данные с замененными макронами (вал тест тоже)

```python
val_with_macrons = full_data_preproc(
    old_val,
    change_macrons=True,
    cols2change=['rus', 'mansi']
)

test_with_macrons = full_data_preproc(
    old_test,
    change_macrons=True,
    cols2change=['rus', 'mansi']
)
```

```python
full_train_with_macrons.to_csv('cleared_v3.1/big_macrons_changed/cleared_v3_train_09.csv', index=False)
val_with_macrons.to_csv('cleared_v3.1/big_macrons_changed/cleared_v3_val_005.csv', index=False)
test_with_macrons.to_csv('cleared_v3.1/big_macrons_changed/cleared_v3_test_005.csv', index=False)
```
